```python
import os

image_files = []

for filename in os.listdir('/content/drive/MyDrive/500_Instances/Data/'):
    if filename.endswith(".jpg"):
        image_files.append("/content/drive/MyDrive/500_Instances/Data/" + filename)
os.chdir("..")
with open("/content/train.txt", "w") as outfile:
    for image in image_files:
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
```
